```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
weights= np.random.normal(loc=65,scale=5,size=280000000)
```

```python
weights
min_weight = np.min(weights)
max_weight = np.max(weights)
print(min_weight)
print(max_weight)
bin_width = (max_weight - min_weight)/500
print(bin_width)
```

```python
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
axes = plt.hist(weights,bins=500)
```

```python
z_score = (weights - np.mean(weights)) / np.std(weights)
```

```python
plt.xlabel("Z-score of weight")
plt.ylabel("Probability (Z-score of weight)")
plt.bar(axes[1][1:],axes[0]/10000000)
```
